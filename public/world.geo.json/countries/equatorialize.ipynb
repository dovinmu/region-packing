```python
import json
import os
with open('../../centroids.json') as f:
    centroids = json.loads(f.read())
centroids['USA']
```

```python
def equatorialize(code):
    cent = centroids[code]
    d3_call = f'"d3.geoOrthographic().rotate([{-cent[0]}, {-cent[1]}])"'
    call = f'geoproject {d3_call} < {code}.geo.json > {code}.ortho.geo.json'
    print(call)
    out = os.system(call)
    # > {code}.ortho.geo.json'
    print(code, out)
```

```python
equatorialize('USA')
```

```python
import os
for fname in os.listdir():
    print(fname)
    code = fname.split('.')[0]
    if code not in ['', 'equatorial', 'equatorialize', 'get all names']:
        equatorialize(code)
```
